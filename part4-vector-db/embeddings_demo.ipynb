```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

sentences = [
"Virat Kohli scored a century in the cricket match",
"The bowler delivered a fast yorker",
"The batsman hit the ball for six",
"Add salt and pepper while cooking pasta",
"Bake the cake at 180 degrees",
"Chop vegetables before frying",
"Cyber attacks are increasing globally",
"Strong passwords prevent hacking",
"Two factor authentication improves security",
"Install antivirus software to protect systems"
]

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)

similarity = cosine_similarity(embeddings)

sns.heatmap(similarity)
plt.show()
```
